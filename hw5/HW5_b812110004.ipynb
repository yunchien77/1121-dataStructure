```python
import re

# Node class 
class Hnode:
    def __init__(self, key, item):
        self.key = key
        self.item = item
        self.parent = None
        self.right = None
        self.left = None

    def getKey(self):
        return self.key

    def getItem(self):
        return self.item

    def getRightChild(self):
        return self.right

    def getLeftChild(self):
        return self.left

    def getParent(self):
        return self.parent

    def hasRightChild(self):
        return (self.right!=None)
 
    def hasLeftChild(self):
        return (self.left!=None)

    def isRoot(self):
        return (self.parent==None)

    def isLeaf(self):
        return ((self.right==None)and(self.left==None))

    def isRightChild(self):
        return (self.parent.right==self)

    def isLeftChild(self):
        return (self.parent.left==self)

    def setParent(self, p):
        self.parent=p

    def setKey(self, key):
        self.key = key

    def setItem(self, item):
        self.item = item
        
    def addRightChild(self, hnode):
        self.right=hnode
        hnode.setParent(self)

    def addLeftChild(self, hnode):
        self.left=hnode
        hnode.setParent(self)

# Heap class 
class Heap:
    def __init__(self):
        self.root=None
        self.last=None
        self.size=0

    def isEmpty(self):
        return (self.size==0)

    def getSize(self):
        return self.size

    def getRoot(self):
        return self.root

    def getLast(self):
        return self.last

    def getHighestPriority(self):
        return self.root.getKey()

    def setRoot(self, hnode):
        self.root=hnode

    def setLast(self, hnode):
        self.last=hnode    


#
# methods for heaps
#

    #
    # downward adjustment from current node
    # 
    def downwardHeapify(self, current):
        # 當堆為空時
        if self.size == 0:
            self.root = None
            print("The heap is now empty")
            return

        # 當堆僅包含一個節點時
        if self.size == 1:
            return

        # 如果當前節點的左子節點不存在，右子節點存在
        if current.left == None and current.right != None:
            if current.key > current.right.key:
                # 存儲右子節點的參考
                temp = current.right
                current.left = None
                current.right = None
                current.parent = temp

                # 重新安排節點結構
                temp.parent = None
                temp.right = current
                temp.left = None
                self.root = temp
            return

        # 如果當前節點的右子節點不存在，左子節點存在
        elif current.right == None and current.left != None:
            if current.key > current.left.key:
                # 存儲左子節點的參考
                temp = current.left
                current.left = None
                current.right = None
                current.parent = temp

                # 重新安排節點結構
                temp.parent = None
                temp.right = None
                temp.left = current
                self.root = temp
            return

        while True:
            # 如果當前節點是根節點
            if current == self.root:
                # 如果當前節點比左子節點和右子節點都小，堆的結構已經正確
                if current.key < current.left.key and current.key < current.right.key:
                    return

                min_child = None
                mid_child = None
                direction = None

                # 如果左右子節點有一個不存在，則退出迴圈
                if current.left == None or current.right == None:
                    return
                # 比較左右子節點，找到較小的那個作為新的根節點
                elif current.left.key < current.right.key:
                    min_child = current.left
                    direction = "left"
                    mid_child = current.right
                elif current.right.key < current.left.key:
                    min_child = current.right
                    direction = "right"
                    mid_child = current.left

                min_child_left = min_child.left
                min_child_right = min_child.right

                # 將最小的子節點提升為新的根節點
                min_child.parent = None
                self.root = min_child

                if direction == "left":
                    min_child.left = current
                    min_child.right = mid_child

                if direction == "right":
                    min_child.right = current
                    min_child.left = mid_child

                if mid_child != None:
                    mid_child.parent = min_child

                current.left = min_child_left
                current.right = min_child_right
                current.parent = min_child

                if min_child_left != None:
                    min_child_left.parent = current

                if min_child_right != None:
                    min_child_right.parent = current
            else:
                if current.left == None and current.right == None:
                    return

                if (
                    current.right != None
                    and current.key < current.left.key
                    and current.key < current.right.key
                ):
                    return

                # 存儲當前節點的父節點以及與當前節點相關的子節點
                temp = current.parent
                min_child = None
                mid_child = None
                direction = None

                if (
                    current.left != None
                    and current.right == None
                    and (current.key < current.left.key)
                ):
                    return

                if (
                    current.left != None
                    and current.right == None
                    and (current.key > current.left.key)
                ):
                    min_child = current.left
                    direction = "left"
                    mid_child = current.right
                elif current.left.key < current.right.key:
                    min_child = current.left
                    direction = "left"
                    mid_child = current.right
                elif current.right.key < current.left.key:
                    min_child = current.right
                    direction = "right"
                    mid_child = current.left

                min_child_left = min_child.left
                min_child_right = min_child.right

                # 更新父節點的左或右子節點為較小的子節點
                if temp.left == current:
                    temp.left = min_child

                if temp.right == current:
                    temp.right = min_child

                min_child.parent = temp

                if direction == "left":
                    min_child.left = current
                    min_child.right = mid_child

                if direction == "right":
                    min_child.right = current
                    min_child.left = mid_child

                if mid_child != None:
                    mid_child.parent = min_child

                current.left = min_child_left
                current.right = min_child_right
                current.parent = min_child

                if min_child_left != None:
                    min_child_left.parent = current

                if min_child_right != None:
                    min_child_right.parent = current



    #
    # upward adjustment toward the root from the current node
    # 
    def upwardHeapify(self, current):
        # 當當前節點的鍵值小於父節點的鍵值時，持續向上調整
        while current.key < current.parent.key:
            # 如果父節點的父節點不存在，即當前節點已經是根節點
            if current.parent.parent == None:
                direction = None
                temp = current.parent  # 暫存父節點
                mid_child = None  # 中間節點
                min_left_child = current.left  # 左子節點
                min_right_child = current.right  # 右子節點

                # 確定當前節點在父節點的左邊還是右邊
                if current.parent.left == current:
                    direction = "left"
                    mid_child = current.parent.right
                elif current.parent.right == current:
                    direction = "right"
                    mid_child = current.parent.left

                # 重新安排節點間的連接關係，以進行向上調整
                temp.left = min_left_child
                temp.right = min_right_child
                temp.parent = current

                # 更新子節點的父節點為暫存的父節點
                if min_left_child != None:
                    min_left_child.parent = temp

                if min_right_child != None:
                    min_right_child.parent = temp

                # 調整節點的連接方向，使得父節點置於當前節點的左邊或右邊
                if direction == "left":
                    current.left = temp
                    current.right = mid_child

                if direction == "right":
                    current.right = temp
                    current.left = mid_child

                # 更新中間節點的父節點為當前節點
                if mid_child != None:
                    mid_child.parent = current

                # 將當前節點的父節點設為空，表示當前節點成為新的根節點
                current.parent = None

                # 更新堆的根節點為當前節點
                self.root = current
                break

            else:
                direction1 = None  
                direction2 = None  
                temp = current.parent  # 暫存父節點  
                temp_parent = current.parent.parent  # 暫存父節點的父節點  
                mid_child = None  # 中間節點
                min_left_child = current.left  # 左子節點
                min_right_child = current.right  # 右子節點

                # 確定當前節點在父節點的左邊還是右邊
                if current.parent.left == current:
                    direction1 = "left"
                    mid_child = current.parent.right
                elif current.parent.right == current:
                    direction1 = "right"
                    mid_child = current.parent.left

                # 確定父節點在其父節點的左邊還是右邊
                if current.parent.parent.left == current.parent:
                    direction2 = "left"
                elif current.parent.parent.right == current.parent:
                    direction2 = "right"

                # 重新安排節點間的連接關係，以進行向上調整
                temp.left = min_left_child
                temp.right = min_right_child
                temp.parent = current

                # 更新子節點的父節點為暫存的父節點
                if min_left_child != None:
                    min_left_child.parent = temp

                if min_right_child != None:
                    min_right_child.parent = temp

                # 調整節點的連接方向，使得父節點置於當前節點的左邊或右邊
                if direction1 == "left":
                    current.left = temp
                    current.right = mid_child

                if direction1 == "right":
                    current.right = temp
                    current.left = mid_child

                # 更新當前節點的父節點為其父節點的父節點
                current.parent = temp_parent

                # 更新中間節點的父節點為當前節點
                if mid_child != None:
                    mid_child.parent = current

                # 將當前節點放入其父節點的位置，繼續向上調整
                if direction2 == "left":
                    temp_parent.left = current
                elif direction2 == "right":
                    temp_parent.right = current




    #
    # The most important operation for a heap, remove_Min (extract_Min())
    #
    def removeMin(self):
        if self.root is None:
            print("The heap is empty and no entry can be removed")
            return

        self.size -= 1

        # 找到最後一個節點
        queue = [self.root]
        last_node = None

        while queue:
            last_node = queue.pop(0)  # 從隊列中移除節點
            # 將左右子節點加入隊列
            if last_node.left is not None:
                queue.append(last_node.left)
            if last_node.right is not None:
                queue.append(last_node.right)

        # 將最後一個節點的值替換到根節點
        self.root.key = last_node.key
        self.root.item = last_node.item

        # 刪除最後一個節點
        if last_node.parent is not None:
            # 根據最後一個節點是其父節點的左邊還是右邊，來設置父節點的左/右子節點為 None
            if last_node.parent.left == last_node:
                last_node.parent.left = None
            else:
                last_node.parent.right = None

        # 調整堆的結構
        self.downwardHeapify(self.root)




    #
    # Another important operation for a heap, insertion() (add())
    #
    def Insert(self, hnode):
        # 如果堆是空的，將根節點設為新節點，並增加堆大小
        if self.size == 0:
            self.root = hnode
            self.size += 1
            return

        # 增加堆大小
        self.size += 1

        # 創建一個佇列，從根節點開始尋找空位
        queue = [self.root]

        while True:
            current = queue[0]
            queue.pop(0)

            # 如果左子節點是空的，將新節點插入左子節點位置，並設定父節點為當前節點
            if current.left is None:
                current.left = hnode
                hnode.parent = current
                break

            # 如果右子節點是空的，將新節點插入右子節點位置，並設定父節點為當前節點
            if current.right is None:
                current.right = hnode
                hnode.parent = current
                break

            # 如果左子節點不是空的，將左子節點加入佇列中
            if current.left is not None:
                queue.append(current.left)

            # 如果右子節點不是空的，將右子節點加入佇列中
            if current.right is not None:
                queue.append(current.right)

        # 對新插入的節點進行向上調整，以維持堆的性質
        self.upwardHeapify(hnode)


    #
    # For management, print the heap in pre-order
    #
    def printHeapPreOrder(self, i):
        # 如果節點是葉子節點，只有左右節點皆為 None，則印出該葉子節點的資訊
        if i != None and i.left == None and i.right == None:
            print(f"Leaf [ {i.key}  {i.item} ]")

            # 若存在左節點，繼續向左遞迴印出
            if i.left != None:
                self.printHeapPreOrder(i.left)

            # 若存在右節點，繼續向右遞迴印出
            if i.right != None:
                self.printHeapPreOrder(i.right)
        elif i != None:
            # 若不是葉子節點，則印出該節點為內部節點
            print(f"Node [ {i.key}  {i.item} ]")

            # 繼續向左遞迴印出左節點
            if i.left != None:
                self.printHeapPreOrder(i.left)

            # 繼續向右遞迴印出右節點
            if i.right != None:
                self.printHeapPreOrder(i.right)



# decimal to binary using yield
def d2by(x):
    # 當輸入的十進制數字大於 0 時執行循環
    while x > 0:
        # 使用 divmod 函數將輸入的十進制數字除以 2，得到商和餘數
        x, r = divmod(x, 2)
        # 使用 yield 返回二進制數字的每一位
        yield r


# function for starting the task
def HeapwithEntriesInserted():
    #
    # read the input information from the default input text file into an
    # entry list, entry_list
    #
    
    #
    # initiating a heap object h
    #
    h=Heap()

    #
    # Do something here to build the heap
    #
    # 開啟預設的輸入文字檔案 "inFile.txt"
    with open("inFile-1.txt", mode="r") as file:
        # 逐行讀取檔案內容
        while True:
            line = file.readline()
            # 如果已經讀取到檔案的最後一行，停止迴圈
            if not line:
                break
            else:
                # 移除每一行的換行符號
                if "\n" in line:
                    line = line[:line.index("\n")]
                # 使用空格分割每一行的數字和文字部分
                arr = re.split(" ", line)
                # 將分割後的數字和文字分別傳遞給 Hnode 類別，建立新節點，並將其插入堆中
                h.Insert(Hnode(int(arr[0]), arr[1]))


                
                

    #---------------------Print as the example on the homework sheet --------
    #                       will be adapted with the input
    #
    print("Heap size=", h.getSize(), "The highest priority is ", h.getHighestPriority())
    print("pre-order traversal:")
    h.printHeapPreOrder(h.getRoot())

    print("deleteMin")
    h.removeMin()

    print("deleteMin")
    h.removeMin()

    print("deleteMin")
    h.removeMin()

    print("deleteMin")
    h.removeMin()

    print("deleteMin")
    h.removeMin()

    print("insert 35, resume")
    h.Insert(Hnode(35, "resume"))
    
    print("insert 15, second")
    h.Insert(Hnode(15, "second"))

    print("insert 20, fourth")
    h.Insert(Hnode(20, "fourth"))

    print("Heap size=", h.getSize(), "The highest priority is ", h.getHighestPriority())
    print("pre-order traversal:")
    h.printHeapPreOrder(h.getRoot())
    
    print("deleteMin")
    h.removeMin()

    print("insert 40, nineth")
    h.Insert(Hnode(40, "nineth"))

HeapwithEntriesInserted()
```
